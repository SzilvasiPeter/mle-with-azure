# Visualize the Training History

Plot the training and validation accuracy to check if there is over or under fitting.

```python
import json
import matplotlib.pyplot as plt

with open("../logs/mnist_history.json", "r") as file:
    history = json.load(file)

with open("../logs/mnist_result.json", "r") as file:
    test_accuracy = json.load(file)["test_accuracy"]

# Plot the training history using matplotlib
plt.plot(history['accuracy'], label='train accuracy')
plt.plot(history['val_accuracy'], label='val accuracy')
plt.axhline(y=test_accuracy, color='r', linestyle='--', label='test accuracy')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.legend()
plt.title('Training History')
plt.show()
```
